```python
import Credentials
import tweepy
import pandas as pd
import os
import time
```

```python
api_key = Credentials.api_key
secret =  Credentials.secret
#bearer_token = ''
access_token = Credentials.access_token
access_token_secret = Credentials.access_token_secret
```

```python
#### v1.1 with Tweepy
auth = tweepy.OAuth1UserHandler(api_key, secret, access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
#now we are getting last 3 recent corona tags 
def hashtags(q,lang,count):
    r = api.search(q,lang,count,result_type="recent")
    return r
```

```python
try:
    #The number of tweets we want to retrieved from the search
    tweets = api.search_tweets(q="#FCK", count=150)
    
    #Pulling Some attributes from the tweet
    attributes_container = [[tweet.user.name, tweet.created_at, tweet.favorite_count, tweet.source,  tweet.text,tweet.hashtags] for tweet in tweets]

    #Creation of column list to rename the columns in the dataframe
    columns = ["User", "Date Created", "Number of Likes", "Source of Tweet", "Tweet","hashtags"]
    
    #Creation of Dataframe
    tweets_df = pd.DataFrame(attributes_container, columns=columns)
except BaseException as e:
    print('Status Failed On,',str(e))
```
